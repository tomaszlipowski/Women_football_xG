# Football match panel

This dashboard presents a summary of the chosen match from statsbombpy, showcasing key match elements:

### Components:

- **Pass Networks**:

    -  The lines indicate passing connections between players who completed **at least 4 passes** to each other.
    -  Displays data only up to the first substitution in the team, but not earlier than halftime.

- **Central section displays match statistics**:
    - xG (Expected Goals)
    - Shots
    - Shots on Target
    - Passes
    - Pass Completion Percentage


- **Shot Maps**:
    - The colorful circles represent **goals**, and their size corresponds to the **xG value** of each shot (larger circles indicate higher xG).
    - This gives insight into the quality of chances created by each team.


- **xG Timeline**:
    - The timeline below shows the accumulation of **xG over time**, marking goal scorers.

### Customization:

Creating a similar dashboard for another match requires very little effort:
1. Find the unique **`match_id`** for the specific game.
2. Customize the **team logos** and **colors** for an enhanced visual experience. (Optional)

```python
df = sb.events(match_id).sort_values(by = ['index'])

# Removes penalty shootouts if needed
df = df[df['period'] < 5] 
```

```python
# Splits the data into two teams

TEAM1_NAME, TEAM2_NAME = df['team'].unique()[0], df['team'].unique()[1]

TEAM1 = df[df.team == TEAM1_NAME]
TEAM2 = df[df.team == TEAM2_NAME]
```

```python
# Creates shotmaps

def create_shotmap(team, ax, team_color):
    
    pitch = VerticalPitch(pitch_type = 'statsbomb', half = True)
    pitch.draw(ax = ax)
    
    team = team[team['type'] == 'Shot']
    
    team = team[['team', 'player', 'location', 'shot_outcome', 'shot_statsbomb_xg', 'shot_type']].copy()
    team[['x', 'y']] = pd.DataFrame(team['location'].tolist(), index = team.index)
    
    # Draws the shotmap
    for shot in team.to_dict(orient = 'records'):
    
        pitch.scatter(
            x = shot['x'], y = shot['y'], ax = ax, 
            s = 500 * shot['shot_statsbomb_xg'], 
            color = team_color if shot['shot_outcome'] == 'Goal' else 'white', 
            edgecolors = 'black', alpha = 0.7
        )
```

```python
# Creates passnetworks

def create_passnetwork(team, ax, team_color):

    team = team.copy()
    team[['x', 'y']] = team['location'].apply(pd.Series)

    # Takes jersey numbers
    tactics = team[~team['tactics'].isnull()][['tactics', 'team']]

    all_jersey_numbers = []
    for y in tactics.to_dict(orient='records'):
        all_jersey_numbers.append(pd.DataFrame([{'player_id': x['player']['id'], 'jersey_number': x['jersey_number']} for x in y['tactics']['lineup']]))
        
    jersey_numbers = pd.concat(all_jersey_numbers).drop_duplicates()
    
    # Makes a new, single column for time and sort the events in chronological order
    team = team.copy()
    team.loc[:, 'newsecond'] = 60 * team['minute'] + team['second']
    team.sort_values(by = ['newsecond'])
    
    # Identifies the passer and then the recipient, who will be the player_Id of the next action
    team['passer'] = team['player_id']
    
    team['recipient'] = team['passer'].shift(-1)
    
    passes_df = team.loc[team['type'] == "Pass"].copy()

    # Filters for only passes and then successful passes
    passes_df.loc[:, 'pass_outcome'] = passes_df['pass_outcome'].fillna("Successful")
    
    # Filerts only successful passes
    completions = passes_df.loc[passes_df['pass_outcome'] == "Successful"]
    
    # Finds time of the team's first substitution and filter the df to only passes before that
    sub_df = team.loc[(team['type'] == "Substitution")]
    first_sub = sub_df["newsecond"].min()
    
    if first_sub <= (60 * 45):
        first_sub = 60 * 45
    
    completions = completions.loc[completions['newsecond'] < first_sub]
    
    # Finds average locations
    average_locs_and_count = completions.groupby('passer').agg({'x': ['mean'], 'y': ['mean','count']})
    average_locs_and_count.columns = ['x', 'y', 'count']
    
    # Finds number of passes along each 'path' of passer to recipient
    passes_between = completions.groupby(['passer', 'recipient']).id.count().reset_index()
    passes_between.rename({'id': 'pass_count'}, axis = 'columns', inplace = True)
    
    passes_between = passes_between.merge(average_locs_and_count, left_on = 'passer', right_index = True)
    passes_between = passes_between.merge(
        average_locs_and_count, left_on = 'recipient', right_index = True, suffixes = ['', '_end']
    )
    
    # Sets threshold for pass arrows to be plotted.
    passes_between = passes_between.loc[(passes_between['pass_count'] >= 4)]
    
    # Plots arrows
    def pass_line_template(ax, x, y, end_x, end_y, line_color):
        
        ax.annotate(
            '',
            xy = (end_y,  end_x),
            xytext = (y, x),
            zorder = 1,
            arrowprops = dict(arrowstyle='-|>', linewidth = 4, color = line_color, alpha = 0.85)
        )
        
    def pass_line_template_shrink(ax, x, y, end_x, end_y, line_color, dist_delta=1.2):
        
        dist = math.hypot(end_x - x, end_y - y)
        angle = math.atan2(end_y-y, end_x-x)
        upd_x = x + (dist - dist_delta) * math.cos(angle)
        upd_y = y + (dist - dist_delta) * math.sin(angle)
        pass_line_template(ax, x, y, upd_x, upd_y, line_color=line_color)
    
    pitch = VerticalPitch(pitch_type = 'statsbomb')
    pitch.draw(ax=ax)
    
    for index, row in passes_between.iterrows():
        pass_line_template_shrink(ax, row['x'], row['y'], row['x_end'], row['y_end'], 'black')
    
    # Plots nodes
    pitch.scatter(
        average_locs_and_count.x, average_locs_and_count.y, s=500,
        color = '#f0ece2', edgecolors = team_color, linewidth = 2, alpha = 1, ax = ax, zorder = 2
    )
    
    for index, row in average_locs_and_count.iterrows():
        
        pitch.annotate(
            jersey_numbers[jersey_numbers['player_id'] == row.name]['jersey_number'].values[0],
            xy = (row.x, row.y),
            c = '#132743',
            va = 'center',
            ha = 'center',
            size = 10,
            fontweight = 'bold',
            ax = ax
        )
```

```python
# Creates tables

def create_table(team1, team2):
    
    # Drops rows where the pass_outcome is Unknown
    team1 = team1[team1['pass_outcome'] != 'Unknown']
    team2 = team2[team2['pass_outcome'] != 'Unknown']
    
    # Create a dictionary to store the stats
    stats = {
        'xG': {'team1': round(team1['shot_statsbomb_xg'].sum(), 2), 'team2': round(team2['shot_statsbomb_xg'].sum(), 2)},
        'shots': {'team1': len(team1[team1['type'] == 'Shot']), 'team2': len(team2[team2['type'] == 'Shot'])},
        'shots_on_target': {
            'team1': len(team1[(team1['type'] == 'Shot') & (team1['shot_end_location'].str.len() > 2) & (team1['shot_outcome'] != 'Off T')]), 
            'team2': len(team2[(team2['type'] == 'Shot') & (team2['shot_end_location'].str.len() > 2) & (team2['shot_outcome'] != 'Off T')])
        },
        'passes': {'team1': len(team1[team1['type'] == 'Pass']), 'team2': len(team2[team2['type'] == 'Pass'])},
        'pass_completion': {
            'team1': round(len(team1[(team1['type'] == 'Pass') & (team1['pass_outcome'].isna())]) / len(team1[team1['type'] == 'Pass']) * 100, 2) ,
            'team2': round(len(team2[(team2['type'] == 'Pass') & (team2['pass_outcome'].isna())]) / len(team2[team2['type'] == 'Pass']) * 100, 2)
        }
    }
    
    stats_df = pd.DataFrame(stats).T

    column_labels = ['', '', '']
    table_vals = [
        [str(stats_df['team1']['xG']), 'xG', str(stats_df['team2']['xG'])],
        [str(int(stats_df['team1']['shots'])), 'Shots', str(int(stats_df['team2']['shots']))],
        [str(int(stats_df['team1']['shots_on_target'])), 'Shots on Target', str(int(stats_df['team2']['shots_on_target']))],
        [str(int(stats_df['team1']['passes'])), 'Passes', str(int(stats_df['team2']['passes']))],
        [f"{stats_df['team1']['pass_completion']}%", 'Pass Completion', f"{str(stats_df['team2']['pass_completion'])}%"]
    ]
    
    return column_labels, table_vals
```

```python
# Creates xG flow charts

def create_xg_flow_chart(df, ax):

    home_team_id = df['team_id'].unique()[0]
    
    # Filters columns to only shots
    df = df[df['type'] == 'Shot']
    
    # Sorts the values so they are in order of time
    df = df.sort_values(by = "index")
    
    # Calculates the cumulative xG for each team
    df['cumulative_xG'] = df.groupby('team_id')['shot_statsbomb_xg'].cumsum()
    
    # Plots the xG for each team by half
    for team in df['team_id'].unique():
        
        team_df = df[df['team_id'] == team]
        
        # add a 0 xG row at the start of the match
        team_df = pd.concat([pd.DataFrame({
            'team_id': team, 'minute': 0, 'shot_statsbomb_xg': 0, 'shot_outcome': None, 'cumulative_xG': 0, 'period': 1
        }, index=[0]), team_df])
        
        ax.plot(
            team_df['minute'], 
            team_df['cumulative_xG'], 
            label = team, 
            drawstyle = 'steps-post',
            c = TEAM1_COLOR if team == home_team_id else TEAM2_COLOR
        )   
            
    
    # Adds a scatter plots to show the goals
    for team in df['team_id'].unique():
        
        team_df = df[(df['team_id'] == team) & (df['shot_outcome'] == 'Goal')].to_dict(orient = 'records')
        
        for x in team_df:

            # Adds circles for goals
            ax.scatter(
                x['minute'], 
                x['cumulative_xG'], 
                c = TEAM1_COLOR if team == home_team_id else TEAM2_COLOR,
                edgecolor = TEAM1_COLOR if team == home_team_id else TEAM2_COLOR,
                s = 100,
                zorder = 5
            )
            
            # Adds labels for goal scorers
            ax.text(
                x['minute'], 
                x['cumulative_xG'] - 0.07, 
                x['player'], 
                ha = 'center', 
                va = 'center', 
                fontfamily = 'monospace', 
                fontsize = 8,
                zorder = 10
            )
            
    # Labels the x axis with first and second half
    ax.set_xticks([0, 45, 90, 120])
    ax.set_xticklabels(['0\'', '45\'', '90\'', '120\''])
    
    # Gets rid of the right and top spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
```

```python
# Sets up the figure and grid layout

fig = plt.figure(figsize = (15, 12))

ax1 = fig.add_axes([0.225, 0.775, 0.15, 0.15])

ax1.imshow(TEAM1_IMG)
ax1.axis('off')

ax2 = fig.add_axes([0.5, 0.8, 0.3, 0.1])

team_1_text = ax2.text(0.47, 0.4, TEAM1_NAME, fontsize = 24, ha = 'right', fontdict=  {'family': 'monospace'})
team_1_text.set_bbox(dict(facecolor = TEAM1_COLOR, alpha = 0.5, edgecolor = '#43A1D5', boxstyle = 'round'))

team_2_text = ax2.text(0.53, 0.4, TEAM2_NAME, fontsize = 24, ha = 'left', fontdict = {'family': 'monospace'})
team_2_text.set_bbox(dict(facecolor = TEAM2_COLOR, alpha = 0.5, edgecolor = '#002654', boxstyle = 'round'))

# Calculates number of goals
team1_goals_number = len(TEAM1[TEAM1['shot_outcome'] == 'Goal'])
team2_goals_number = len(TEAM2[TEAM2['shot_outcome'] == 'Goal'])

ax2.text(0.47, 0, str(team1_goals_number), fontsize = 20, ha = 'right', fontdict = {'family': 'monospace', 'weight': 'bold'})
ax2.text(0.5, 0, '-', fontsize = 20, ha = 'center', fontdict = {'family': 'monospace', 'weight': 'bold'})
ax2.text(0.53, 0, str(team2_goals_number), fontsize = 20, ha = 'left', fontdict = {'family': 'monospace', 'weight': 'bold'})

ax2.axis('off')

ax3 = fig.add_axes([0.925, 0.775, 0.15, 0.15])

ax3.imshow(TEAM2_IMG)
ax3.axis('off')

ax4 = fig.add_axes([0.15, 0.25, 0.3, 0.5])
create_passnetwork(TEAM1, ax4, TEAM1_COLOR)
ax4.axis('off')

ax5 = fig.add_axes([0.5, 0.25, 0.3, 0.5])

# creates a table with data.
column_labels, table_vals = create_table(TEAM1, TEAM2)

table = ax5.table(
    cellText = table_vals, 
    cellLoc = 'center', 
    edges = 'vertical',
    bbox = [0, 0.5, 1, 0.45],
)

table.set_fontsize(14)

# Displays the appropriate table borders
for (i, j), cell in table.get_celld().items():
    if j == 0:
        table.get_celld()[(i, j)].visible_edges = 'R'
    elif j == 2:
        table.get_celld()[(i, j)].visible_edges = 'L'
    else:
        table.get_celld()[(i, j)].visible_edges = 'LR'

ax5.axis('off')

ax6 = fig.add_axes([0.85, 0.25, 0.3, 0.5])
create_passnetwork(TEAM2, ax6, TEAM2_COLOR)

ax7 = fig.add_axes([0.2, 0.05, 0.2, 0.2])
create_shotmap(TEAM1, ax7, TEAM1_COLOR)

ax8 = fig.add_axes([0.5, 0.05, 0.35, 0.35])
create_xg_flow_chart(df, ax8)

ax9 = fig.add_axes([0.9, 0.05, 0.2, 0.2])
create_shotmap(TEAM2, ax9, TEAM2_COLOR)
ax9.axis('off')
```
